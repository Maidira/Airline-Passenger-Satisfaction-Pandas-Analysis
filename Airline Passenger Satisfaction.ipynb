## Exploratory data analysis

```python
# Satisfication

df['satisfaction'].value_counts(normalize = True)
```

```python
sns.countplot(data = df, x = 'satisfaction')
plt.title('Passenger satisfaction distribution')
plt.show()
```

```python
# Class vs Satisfication
sns.countplot(data = df, x = 'Class', hue = 'satisfaction')
plt.title('Satisfaction by Class')
plt.show()
```

```python
# Flight Distance vs Satisfaction
sns.histplot(data = df, x = 'Flight Distance', hue = 'satisfaction', kde = True, bins = 30)
plt.title('Flight Distance vs Satisfaction')
plt.show()
```

```python
# Satisfaction by Age Group
bins = [0,18 ,25, 40, 60 , 80]
labels = ['1-18','19-25', '26-40', '41-60' ,'60+']
df['age_group'] = pd.cut(df['Age'], bins = bins, labels = labels)

sns.countplot(data = df, x= 'age_group', hue = 'satisfaction')
plt.title('Satisfaction by Age')
plt.show()
```

```python
# Avg rating by Satisfaction
rating_cols = [col for col in df.columns if 'service' in col or 'onboard' in col or 'baggage' in col or 'Cleanliness' in col or 'Leg room service' in col or 'Seat comfort' ]


df.groupby('satisfaction')[rating_cols].mean()
```

```python
# Satisfaction by class and Type of travel
pd.pivot_table(df, index='Class', columns='Type of Travel', values='satisfaction', aggfunc=lambda x: (x=='satisfied').mean())
```

```python
# Average Flight Distance by Airline Status
df.pivot_table(index='satisfaction', columns='Class', values='Flight Distance', aggfunc='mean')
```

```python
# Correlation Heatmap
numeric_cols = df.select_dtypes(include='number')
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

### KPI's

```python
# Net promoter score
df['nps_group'] = df['satisfaction'].apply(lambda x: 'Promoter' if x == 'satisfied' else 'Detractor')

nps = df['nps_group'].value_counts(normalize=True)
nps_score = round((nps['Promoter'] - nps['Detractor']) * 100, 2)

print(f"NPS Proxy Score: {nps_score}")
```

```python
# Flight Distance bucket
df['distance_bucket'] = pd.cut(df['Flight Distance'], 
                             bins=[0, 500, 1500, 3000, 5000], 
                             labels=['Short', 'Medium', 'Long', 'Very Long'])

# Calculate normalized proportions
satisfaction_proportions = (df.groupby(['distance_bucket', 'satisfaction'])
                          .size()
                          .groupby(level=0)
                          .apply(lambda x: x / x.sum())
                          .unstack())


satisfaction_proportions.plot(kind='bar', stacked=True)
plt.title('Satisfaction by Flight Distance')
plt.ylabel('Proportion')
plt.xlabel('Flight Distance Category')


import matplotlib.ticker as mtick
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

plt.show()
```

```python
#  Departure & Arrival Delays vs Satisfaction
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.boxplot(data=df, x='satisfaction', y='Departure Delay in Minutes', ax=axes[0])
axes[0].set_title('Departure Delay vs Satisfaction')

sns.boxplot(data=df, x='satisfaction', y='Arrival Delay in Minutes', ax=axes[1])
axes[1].set_title('Arrival Delay vs Satisfaction')

plt.tight_layout()
plt.show()
```

```python
# Service Ratings vs Satisfaction (Violin Plots)
rating_features = ['Inflight wifi service', 'Online boarding', 'Seat comfort',
                   'Inflight entertainment', 'On-board service', 'Baggage handling']

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))

for ax, feature in zip(axes.flat, rating_features):
    sns.violinplot(data=df, x='satisfaction', y=feature, ax=ax)
    ax.set_title(f'{feature.replace("_", " ").title()} vs Satisfaction')

plt.tight_layout()
plt.show()
```

```python
# Average Ratings by Class and Travel Type
pivot = df.pivot_table(values='Inflight entertainment', 
                       index='Class', 
                       columns='Type of Travel', 
                       aggfunc='mean')

sns.heatmap(pivot, annot=True, cmap='Blues')
plt.title('Avg Inflight Entertainment Rating by Class & Travel Type')
plt.show()
```

```python
# Age Distribution
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Age', hue='satisfaction', fill=True)
plt.title('Age Distribution by Satisfaction')
plt.show()
```

```python
# Satisfaction Crosstab with Gender and Class
pd.crosstab(index=df['Gender'], columns=[df['Class'], df['satisfaction']], normalize='index') * 100
```

```python
# Satisfactin by services
rating_cols = ['Inflight wifi service', 'Online boarding', 'Seat comfort', 
               'Inflight entertainment', 'On-board service', 'Baggage handling']

melted = df.melt(id_vars=['satisfaction'], value_vars=rating_cols, var_name='service_type', value_name='rating')

# Plot average rating per service
sns.barplot(data=melted, x='service_type', y='rating', hue='satisfaction')
plt.xticks(rotation=45)
plt.title("Service Ratings by Satisfaction Level")
plt.show()
```

```python
# Top Features Impacting Satisfaction
service_means = df.groupby('satisfaction')[rating_cols].mean().T
diff = service_means['satisfied'] - service_means['neutral or dissatisfied']
diff.sort_values(ascending=False).plot(kind='bar', title='Service Impact on Satisfaction')
```
